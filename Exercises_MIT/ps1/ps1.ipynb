# Parte A
## Time saving money to buy a Hause

```python
house_projection = pd.DataFrame()
```

```python
total_cost = 500000
annual_salary = 4500*12
month_salary = annual_salary/12
current_savings = 0

portion_down_payment = 0.25
portion_saved = 0.1
r = 0.04


class HouseProjection:
    df = pd.DataFrame(columns=['month','total_cost','portion_down_payment','current_savings','invest_return',
                                         'annual_salary','month_salary','portion_saved'])
    
    def __init__(self, total_cost, annual_salary, current_savings, portion_down_payment, portion_saved, r):
        
        self.df.loc[0, 'month'] = 0
        self.df['total_cost'] = total_cost
        self.df['portion_down_payment'] = total_cost*portion_down_payment
        self.df['current_savings'] = current_savings
        self.df['invest_return'] = current_savings*r/12
        self.df['annual_salary'] = annual_salary
        self.df['month_salary'] = annual_salary/12
        self.df['portion_saved'] = portion_saved*annual_salary/12
        
        
    def project(self):            
        i = 0
        while self.df.loc[i, 'current_savings'] < self.df.loc[0, 'portion_down_payment']:
            i = i + 1         
            self.df.loc[i, 'month'] = i
            self.df.loc[i,'total_cost'] = total_cost
            self.df.loc[i,'portion_down_payment'] = total_cost*portion_down_payment
            self.df.loc[i,'current_savings'] = self.df.loc[i-1, 'current_savings'] + self.df.loc[i-1, 'invest_return'] \
                                    + self.df.loc[i-1, 'portion_saved']
            self.df.loc[i, 'invest_return'] = self.df.loc[i, 'current_savings']*r/12
            self.df.loc[i,'annual_salary'] = annual_salary
            self.df.loc[i,'month_salary'] = annual_salary/12
            self.df.loc[i,'portion_saved'] = portion_saved*annual_salary/12
            
        self.months_down_payment = i
                    
        return self.df
```

```python
house_projection = HouseProjection(total_cost, annual_salary, current_savings, portion_down_payment, portion_saved, r)
```

```python
house_projection.df
```

```python
house_projection.project()
print(f'The number of months needed to adquire the down payment {total_cost*portion_down_payment} is {house_projection.months_down_payment} moths')            
```

# Part C
## Best saving rate

```python
total_cost = 1000000
annual_salary = 300000
month_salary = annual_salary/12
current_savings = 0
months_increassing_period = 6 #each six months the salary increas

fraction_increase_salary = 0.07
portion_down_payment = 0.25
r = 0.04

months_down_payment = 36
```

```python
min_saved = 0.01
max_saved = 0.99

from random import random
from time import time

def ran(min_val, max_val):
    ran = random()
    while (min_val > ran) or (ran > max_val):
        ran = random()
    return ran
```

```python
portion_saved = ran(min_saved, max_saved) 
print(portion_saved)
```

```python
house_projection = HouseProjection(total_cost, annual_salary, current_savings, portion_down_payment, portion_saved, r)
precision = 1
prec = False
steps = 0

ti = time()
elapse = 0


while ((house_projection.project().shape[0]-1) != months_down_payment) or prec == False:
    #print('elapse: ', elapse)
    if elapse > 25:
        print('It is not possible to pay the down payment in three years')
        break
        
    steps = steps + 1
    if (house_projection.project().shape[0]-1) > months_down_payment:
        min_saved = portion_saved
    elif (house_projection.project().shape[0]-1) < months_down_payment:
        max_saved = portion_saved
    else:# (house_projection.project().shape[0]-1) == months_down_payment:
        dif = house_projection.df.loc[months_down_payment, 'current_savings']\
                  - house_projection.df.loc[months_down_payment, 'portion_down_payment']
        if dif <= precision:
            prec = True
            print('aqui', dif)
            break
        else:
            max_saved = portion_saved
            
    portion_saved = ran(min_saved, max_saved) 
    
    house_projection = HouseProjection(total_cost, annual_salary, current_savings, portion_down_payment, portion_saved, r)
    
    te = time()
    elapse = te - ti
    ti = te
    
print('steps', steps)    
#print(house_projection.project().shape[0] - 1)
```
