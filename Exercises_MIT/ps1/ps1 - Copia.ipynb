## Time saving money to buy a Hause

```python
house_projection = pd.DataFrame()
```

```python
total_cost = 500000
annual_salary = 4500*12
month_salary = annual_salary/12
current_savings = 0

portion_down_payment = 0.25
portion_saved = 0.1
r = 0.04


class HouseProjection:
    df = pd.DataFrame(columns=['month','total_cost','portion_down_payment','current_savings','invest_return',
                                         'annual_salary','month_salary','portion_saved'])
    
    def __init__(self, total_cost, annual_salary, current_savings, portion_down_payment, portion_saved, r):
        
        self.df.loc[0, 'month'] = 0
        self.df['total_cost'] = total_cost
        self.df['portion_down_payment'] = total_cost*portion_down_payment
        self.df['current_savings'] = current_savings
        self.df['invest_return'] = current_savings*r/12
        self.df['annual_salary'] = annual_salary
        self.df['month_salary'] = annual_salary/12
        self.df['portion_saved'] = portion_saved*annual_salary/12
        
        
    def project(self, n):
        intern_df = self.df.copy()
        if n <= 0:
            return print('Number of months to project must be a positive integer')
        for i in range(1, n+1):
            #print(intern_df)
            #print(i)
            intern_df.loc[i, 'month'] = i
            intern_df.loc[i,'total_cost'] = total_cost
            intern_df.loc[i,'portion_down_payment'] = total_cost*portion_down_payment
            intern_df.loc[i,'current_savings'] = intern_df.loc[i-1, 'current_savings'] + intern_df.loc[i-1, 'invest_return'] \
                                           + intern_df.loc[i-1, 'portion_saved']
            intern_df.loc[i, 'invest_return'] = intern_df.loc[i, 'current_savings']*r/12
            intern_df.loc[i,'annual_salary'] = annual_salary
            intern_df.loc[i,'month_salary'] = annual_salary/12
            intern_df.loc[i,'portion_saved'] = portion_saved*annual_salary/12

        return intern_df
```

```python
house_projection = HouseProjection(total_cost, annual_salary, current_savings, portion_down_payment, portion_saved, r)
```

```python
def monts_portion_down_payment(hproj):
    intern_df = hproj.df.copy()
    i = 0
    while intern_df.loc[i, 'current_savings'] < intern_df.loc[0, 'portion_down_payment']:
        i = i + 1         
        intern_df.loc[i] = hproj.project(i).loc[i]

    print(f'The number of months needed to adquire the down payment {total_cost*portion_down_payment} is {i} moths')

    return intern_df
```

```python
monts_portion_down_payment(house_projection)
```
